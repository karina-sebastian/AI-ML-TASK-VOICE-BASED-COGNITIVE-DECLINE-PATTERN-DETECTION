Extracting audio features

```python
import os
import whisper
import librosa
import noisereduce as nr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import parselmouth
import textstat
import re
import spacy

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
# 4.1 Base Audio Features
def count_pauses(y, sr, top_db=30):
    intervals = librosa.effects.split(y, top_db=top_db)
    return len(intervals)

def estimate_speech_rate(y, sr):
    duration = librosa.get_duration(y=y, sr=sr)
    mfcc = librosa.feature.mfcc(y=y, sr=sr)
    return mfcc.shape[1] / duration if duration > 0 else 0

def extract_audio_features(path):
    y, sr = librosa.load(path, sr=16000)
    y = nr.reduce_noise(y=y, sr=sr)
    return {
        'duration_sec': librosa.get_duration(y=y, sr=sr),
        'zero_crossing_rate': np.mean(librosa.feature.zero_crossing_rate(y)),
        'spectral_centroid': np.mean(librosa.feature.spectral_centroid(y=y, sr=sr)),
        'spectral_rolloff': np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr)),
        'mfcc_mean': np.mean(librosa.feature.mfcc(y=y, sr=sr).T, axis=0)[0],
        'mfcc_std': np.std(librosa.feature.mfcc(y=y, sr=sr).T, axis=0)[0],
        'rms': np.mean(librosa.feature.rms(y=y)),
        'pause_count': count_pauses(y, sr),
        'speech_rate': estimate_speech_rate(y, sr)
    }

# 4.2 Prosody via Parselmouth
def extract_prosody(path):
    snd = parselmouth.Sound(path)
    pitch = snd.to_pitch()
    f0 = pitch.selected_array['frequency']
    f0 = f0[f0 > 0]
    mean_f0 = f0.mean() if len(f0) else 0
    std_f0  = f0.std()  if len(f0) else 0

    pp = parselmouth.praat.call(snd, "To PointProcess (periodic, cc)", 75, 500)
    jitter  = parselmouth.praat.call(pp, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)
    shimmer = parselmouth.praat.call([snd, pp], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
    hnr_obj = parselmouth.praat.call(snd, "To Harmonicity (cc)", 0.01, 75, 0.1, 1.0)
    hnr_val = parselmouth.praat.call(hnr_obj, "Get mean", 0, 0)

    return {
        'mean_f0': mean_f0,
        'std_f0': std_f0,
        'jitter': jitter,
        'shimmer': shimmer,
        'hnr': hnr_val
    }

# 4.3 Extended Pause Dynamics
def pause_stats(y, sr, top_db=30):
    intervals = librosa.effects.split(y, top_db=top_db)
    silences, prev_end = [], 0
    for start, end in intervals:
        silences.append((start - prev_end)/sr)
        prev_end = end
    total_pause = sum(silences)
    avg_pause   = total_pause / len(silences) if silences else 0
    speech_time = librosa.get_duration(y=y, sr=sr) - total_pause
    return {
        'total_pause_dur': total_pause,
        'avg_pause_dur': avg_pause,
        'speech_pause_ratio': speech_time / total_pause if total_pause>0 else 0
    }

# 4.4 Lexical, Syntactic & Disfluency Metrics
def lexical_syntactic_metrics(text):
    words = re.findall(r"\w+", text)
    total, unique = len(words), len(set(words))
    ttr = unique/total if total else 0
    avg_wlen = sum(len(w) for w in words)/total if total else 0

    doc = nlp(text)
    sent_lens = [len(sent) for sent in doc.sents]
    avg_sent_len = sum(sent_lens)/len(sent_lens) if sent_lens else 0

    return {
        'total_words': total,
        'unique_words': unique,
        'ttr': ttr,
        'avg_word_len': avg_wlen,
        'avg_sent_len': avg_sent_len
    }

def disfluency_metrics(text):
    fs = len(re.findall(r"\b(\w+)[\-\—]\1\b", text.lower()))
    rp = len(re.findall(r"\b(\w+)\s+\1\b", text.lower()))
    return {
        'false_start_count': fs,
        'word_repetition_count': rp
    }

def articulation_rate(text, speech_time):
    sylls = textstat.syllable_count(text)
    return sylls / speech_time if speech_time>0 else 0

def count_fillers(text):
    fillers = {"uh","um","hmm","like","you know"}
    return sum(w in fillers for w in text.lower().split())
```

```python
features = []

for file in audio_files:
    try:
        # audio + NLP
        y, sr = librosa.load(file, sr=16000)
        base_feats = extract_audio_features(file)
        pros = extract_prosody(file)
        ps   = pause_stats(y, sr)
        txt  = transcripts.get(file, "")
        ls   = lexical_syntactic_metrics(txt)
        dfm  = disfluency_metrics(txt)
        art  = articulation_rate(txt, base_feats['duration_sec'] - ps['total_pause_dur'])
        fill = count_fillers(txt)

        # merge all
        feat = {**base_feats, **pros, **ps, **ls, **dfm}
        feat.update({
            'articulation_rate': art,
            'filler_count': fill,
            'transcript': txt,
            'filename': file
        })
        features.append(feat)
    except Exception as e:
        print(f"❌ Error processing {file}: {e}")

df = pd.DataFrame(features).set_index('filename')
df.head()
```

CLUSTERING

```python
X = df.drop(columns=['transcript'])
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X_pca)
df['cluster'] = labels



plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels, cmap='coolwarm', s=100)
for i, fname in enumerate(df.index):
    plt.annotate(fname, (X_pca[i,0], X_pca[i,1]), fontsize=8)
plt.title("Cognitive Speech Pattern Clusters")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
```

This shows two distinct groups blue which are likely the patients at risk.
More pauses, Slower speech, More hesitation markers, Inconsistent sentence completion.

**→ These could be markers of early cognitive impairment.**

while red shows Clear, consistent speech , Fewer pauses and hesitation. Suggesting typical or low-risk cognitive function.
